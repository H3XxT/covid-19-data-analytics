```python
covid_df = pd.read_csv('owid-covid-data.csv', parse_dates=['date'])
covid_df.head()
```

```python
import datetime
```

```python
safe_date = datetime.date.today() + pd.Timedelta('-14 days')
safe_date
```

```python
recent_covid = covid_df[covid_df['date'] >= pd.Timestamp(safe_date)]
recent_covid
```

```python
locations = recent_covid.groupby('location')['new_cases'].sum() <= 10
no_new = set(locations[locations].index)
no_new
```

```python
locations = recent_covid.groupby('location')['new_cases'].std() < 10
low = set(locations[locations].index)
low
```

```python
safe = no_new.union(low)
print(len(safe))
print(safe)
```
